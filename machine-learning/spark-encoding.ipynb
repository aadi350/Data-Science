```python
import numpy as np
from pandas import DataFrame

df = DataFrame({
    'cat1': np.random.choice(['a','b','c'], size=100),
    'cat2': np.random.choice(['A','B', 'C', None], size=[100])
})
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
df = spark.createDataFrame(df)
```

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer

indexer = StringIndexer(inputCols=['cat1', 'cat2'], handleInvalid='keep', outputCols=['cat1_numeric', 'cat2_numeric'])

ohe = OneHotEncoder(inputCols=['cat1_numeric', 'cat2_numeric'],handleInvalid='keep',  outputCols=['cat_1_ohe', 'cat_2_ohe'])

df = indexer.fit(df).transform(df)

df = ohe.fit(df).transform(df)
```
