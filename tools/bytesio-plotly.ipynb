```python
from pathlib import PurePosixPath
import io
import plotly.graph_objects as go
```

```python
import numpy as np

x = np.linspace(0, 10, )
y = x**2

fig = go.Figure(
    go.Scatter(x=x, y=y))
```

```python
bytes_buffer = io.BytesIO()

bytes = io.BytesIO(fig.to_image(format='png'))
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(6, 6))
plt.scatter(x, y)
fig.savefig(bytes_buffer)
```

```python
import fsspec
_fs = fsspec.filesystem('file', )
```

```python
with _fs.open('matplotlib_fig.png', mode='wb') as fs_file:
    fs_file.write(bytes_buffer.getvalue())
```
